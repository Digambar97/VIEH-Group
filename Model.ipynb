```python
import pandas as pd
data = pd.read_csv("/drive/notebooks/CreditCardData.csv")
data.head()
```

```python
data.isnull().sum
data.dropna()
```

```python
import matplotlib.pyplot as plt
x = data['dependents']
y = data['months']
# plt.xlabel['dependents']
# plt.ylabel['months']
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
x = np.array(data['dependents']).reshape(-1, 1)
x.shape
y = np.array(data['months'])
y.shape
```

```python
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.40)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
# print(x_train.shpae)
# print(x_test.shape)
# print(y_train.shpae)
# print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
clf.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

clf = RandomForestClassifier()
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
clf.score(x_test, y_test)
```

```python
from sklearn import metrics

print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(x_train, y_train))
```
